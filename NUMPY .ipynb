Arrays are very frequently used in dta science ,where speed and resource are very importantcollection of homogeneous data

** **ARRAY JOINING** **

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.vstack((a,b))

print(c)
```

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.row_stack((a,b))

print(c)
```

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.hstack((a,b))

print(c)
```

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.column_stack((a,b))

print(c)
```

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.concatenate((a,b))

print(c)
```

```python
import numpy as np

a = np.array([(1,2,3),(4,5,6)])

b = np.array([(7,8,9),(10,11,12)])

c = np.concatenate((a,b),axis = 1)

print(c)
```
